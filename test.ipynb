```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn import datasets

plt.rc('font', family='Malgun Gothic')  # 윈도우 사용자
plt.rc('axes', unicode_minus=False)
```

```python
# 데이터 로드 (iris 데이터셋)
iris = datasets.load_iris()
data = pd.DataFrame(data=iris.data, columns=iris.feature_names)
data['target'] = iris.target
```

```python
plt.figure(figsize=(10, 6))
x = np.arange(len(data))
y = data['sepal length (cm)']
plt.plot(x, y, label='꽃받침 길이', color='b')
plt.xlabel('샘플 인덱스')
plt.ylabel('꽃받침 길이 (cm)')
plt.title('꽃받침 길이의 선 그래프')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
x = data['sepal length (cm)']
y = data['sepal width (cm)']
plt.scatter(x, y, c=data['target'], cmap='viridis')
plt.xlabel('꽃받침 길이 (cm)')
plt.ylabel('꽃받침 너비 (cm)')
plt.title('꽃받침 길이와 너비의 산점도')
plt.colorbar(label='품종 레이블')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
data_means = data.groupby('target').mean()
x = iris.target_names
y = data_means['sepal length (cm)']
plt.bar(x, y, color=['r', 'g', 'b'])
plt.xlabel('품종')
plt.ylabel('평균 꽃받침 길이 (cm)')
plt.title('품종별 평균 꽃받침 길이의 막대 그래프')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(data['petal length (cm)'], bins=20, color='c', edgecolor='black')
plt.xlabel('꽃잎 길이 (cm)')
plt.ylabel('빈도')
plt.title('꽃잎 길이의 히스토그램')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
counts = data['target'].value_counts()
labels = iris.target_names
plt.pie(counts, labels=labels, autopct='%1.1f%%', startangle=90, colors=['lightblue', 'lightgreen', 'lightcoral'])
plt.title('품종별 데이터 비율의 파이 차트')
plt.show()
```

```python
import seaborn as sns
plt.figure(figsize=(10, 6))
corr_matrix = data.iloc[:, :-1].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('특성 간의 상관관계 히트맵')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
data_means = data.groupby('target').mean()
x = iris.target_names
y1 = data_means['sepal length (cm)']
y2 = data_means['petal length (cm)']
plt.fill_between(x, y1, color='skyblue', alpha=0.4, label='꽃받침 길이')
plt.fill_between(x, y2, color='salmon', alpha=0.4, label='꽃잎 길이')
plt.xlabel('품종')
plt.ylabel('평균 측정 값 (cm)')
plt.title('품종별 평균 꽃받침 및 꽃잎 길이의 영역 그래프')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.boxplot([data['sepal length (cm)'], data['sepal width (cm)'],
             data['petal length (cm)'], data['petal width (cm)']],
            labels=iris.feature_names)
plt.xlabel('특성')
plt.ylabel('측정 값 (cm)')
plt.title('Iris 특성의 박스 플롯')
plt.show()
```
